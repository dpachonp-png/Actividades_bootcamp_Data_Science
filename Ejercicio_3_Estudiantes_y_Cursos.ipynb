```python
mates = ["Juan", "Daniel", "Laura", "Sara", "Angelica", "Maira", "Samuel", "Luna"]

ingles = ["Juan", "Laura", "Angelica", "Maira", "Samuel", "Gabriel", "Susana"]
```

```python
ambos = set(mates)&set(ingles)
print(ambos)
```

```python
solo_mates = set(mates)-set(ingles)
print(solo_mates)
solo_ingles = set(ingles)-set(mates)
print(solo_ingles)
```

```python
todos = set(mates).union(set(ingles))
print(todos)
```

```python
todos = list(todos)

for estudiante in todos:
  if estudiante in ambos:
    print ("El estudiante ", estudiante, " esta en ambos cursos")
  elif estudiante in solo_mates:
    print ("El estudiante ", estudiante, " esta solo en el curso de matematicas")
  else:
    print ("El estudiante ", estudiante, " esta solo en el curso de ingles")
```
