```python


clientes = ["Juan", "Daniel", "Laura", "Sara", "Juan", "Laura", "Angelica", "Maira", "Samuel", "Samuel", "Luna", "Luna"]

activos = ["Juan", "Laura", "Angelica"]

inactivos = ["Daniel", "Laura"]
```

```python
sinduplicados = set(clientes)
print (sinduplicados)
```

```python
ambosestados = set(activos).intersection(set(inactivos))
print (ambosestados)
```

```python
soloactivos = 	set(activos).difference(set(inactivos))
print (soloactivos)
```

```python
sinduplicados = list(sinduplicados)
activos = list(activos)
```

```python

for cliente in clientes:
  if cliente in ambosestados:
    print ("El cliente ", cliente, " esta en ambos estados")
  elif cliente in activos:
    print ("El cliente ", cliente, " esta en estado activo")
  else:
    print ("El cliente ", cliente, " esta en estado inactivo")

```
