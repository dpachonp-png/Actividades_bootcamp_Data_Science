### **Parte 2 — Limpieza de datos (Pandas)**

 • Corregir tipos de datos incorrectos

```python
df1["cantidad"]= pd.to_numeric(df1["cantidad"], errors="coerce")
```

- Identificar y manejar valores nulos

```python
df.isnull().sum()
```

```python
# rrelnar valores nulos con el promedio
df1["precio"] = df1["precio"].fillna(df["precio"].median())
```

```python

df1["cantidad"] = df1["cantidad"].fillna(df1["cantidad"].median())
df1.info()
```

- Estandarizar columnas

```python
# Unificar formatos de letra en miniscula y espaciado
df1["pais"]=df1["pais"].str.lower().str.strip()
# Manejar valores con espacios o vacios
df1["pais"] = df1["pais"].replace("", None)
df1["pais"] = df1["pais"].replace(" ", None)
```

```python
#Estandarizar nombres similares para un mismo valor
df1["pais"] = df1["pais"].replace({
    "col" : "colombia",
    "perú" : "peru"
}
)
```

```python
df1["metodo_pago"]=df1["metodo_pago"].str.lower().str.strip()
df1["metodo_pago"] = df1["metodo_pago"].replace("", None)
df1["metodo_pago"] = df1["metodo_pago"].replace(" ", None)
```

• Revisar posibles valores atípicos (outliers)

```python
df1 = df1[df1["precio"]<100000]
```

```python
df1.head()
```

- Corregir formatos de fecha si es necesario

```python
df1["fecha"] = pd.to_datetime(df1["fecha"], errors = "coerce", format = "mixed", dayfirst= True)
df1 = df1.dropna(subset=["fecha"])
```

## **Parte 3 — Análisis con Pandas**

```python
df1['total'] = df1['precio'] * df1['cantidad']
print("Total vendido", df1["total"].sum())
print("Promedio", df1["total"].mean())
print("Venta maxima", df1["total"].max())
print("Venta minima", df1["total"].min())
top_productos = df1.groupby("producto")["cantidad"].sum().sort_values(ascending=False)
print(top_productos)
```

## **Parte 4 — NumPy**

```python
totales = df1[['precio','cantidad']].to_numpy() # convertir columnas numericas a array
# Separar columnas utilizando indexación:
precios = totales[:, 0]
cantidades = totales[:, 1]
```

### **Parte 5 — Análisis con NumPy**

```python
totales = precios * cantidades # CAlculo por vectorizacion
print(totales)
print("total de ventas: ", np.sum(totales))
print("promedio de ventas: ", np.mean(totales))
print("venta minima: ", np.max(totales))
print("ventas mayores a 1000: ", len(totales[totales>1000])) #encontrar valores superiores a 1000 y cuantos hay
```

¿Qué ventajas observas al usar NumPy? Es mas rapido por que usa menos memoria ram, hace operaciones mas complejas ya que usa matrices para acceder a todos los datos de forma mas facil
¿Qué significa “vectorización”? es la tecnica que usa nampy para hacer operaciones a matrices, disminuyendo el uso de ciclos, ya que es capaz de trabajar sobre blouqes para acceder mas rapido a ellos
¿Qué hace la expresión data[:, 0]? Le dice a la matriz que precios la va a encontrar en la columna de precio y que tome la columna completa
