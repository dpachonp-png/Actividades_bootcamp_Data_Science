```python
productos_vendidos = ["Guitarra", "Teclado", "Microfono", "Bajo"]
inventario = ["Bateria", "Bajo", "Amplificador", "Guitarra", "Pedal", "Teclado"]
```

```python
todos = set(productos_vendidos).union(set(inventario))
print (todos)
```

```python
ambos = set(productos_vendidos).intersection(set(inventario))
print(ambos)
```

```python
no_vendidos = set(inventario)-set(productos_vendidos)
print(no_vendidos)
```

```python
i= 0
while i < len(todos):
  for producto in todos:
    if producto in no_vendidos:
      print ("el producto esta a la venta")
    else:
      print("el producto no esta a la venta")
    i=i+1
```
